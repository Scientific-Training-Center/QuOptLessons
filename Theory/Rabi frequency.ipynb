# Rabi frequency

В честь лауреата Нобелевской премии американского физика Исидора Исаака Раби. 

Раби-частота (Rabi frequency) - это частота перехода между состояниями атома или молекулы под воздействием электромагнитного излучения. Она определяется как отношение энергии излучения к энергетическому разности между двумя состояниями. Раби-частота является важным параметром в квантовой оптике и спектроскопии, а также в области квантовой информатики и квантовой компьютерной технологии.

Частота Раби — частота колебаний Раби, существующих в двухуровневых системах (системах, имеющих два переходных состояния все более низкой и высокой энергии) при взаимодействии с околорезонансными движущими полями.

Атом является наименьшей единицей материи, и это классический пример двухуровневой системы (с двумя уровнями) или состояниями: основным состоянием и возбужденным состоянием. Когда свет взаимодействует с атомами, между ними происходит обмен энергией. Этот энергетический обмен носит периодический характер и образует колебания, называемые колебаниями Раби.

Когда энергия, связанная с двумя состояниями атома, различна, тогда атомы будут проходить цикл Раби. Атомы, подвергшиеся воздействию когерентного пучка падающего света, поглощают фотоны циклически. Фотон взаимодействует с электронами в возбужденных атомах и переводит его из возбужденного состояния в основное. Энергия от перешедшего атома испускается обратно в движущее поле в виде нового фотона. Новый фотон движется от атома к падающему свету, вызывая взаимодействие света и атома. Новый фотон обладает теми же параметрами, что и падающий свет — частотой, фазой, поляризацией и ориентацией. Это явление называется циклом Раби, а обратная продолжительность цикла Раби может быть названа частотой Раби.

Применение: Измерение Раби-частоты может использоваться для исследования характеристик квантовых систем, а также для измерения энергетических уровней атомов и молекул. Раби-частота может изменяться в зависимости от типа атома или молекулы, а также от энергии и интенсивности используемого излучения. Раби-частота также играет важную роль в различных квантовых технологиях, таких как квантовая локализация, квантовое кодирование и квантовый компьютер. Однако правило терпит неудачу в неоднородных системах, где жесткость частоты колебаний является результатом изменения частоты движущего поля — в отличие от обобщенного правила. Частотная жесткость в неоднородных системах приводит к спектральному расширению.
